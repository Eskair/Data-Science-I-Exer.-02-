1. Open the file **"ca_sample_data_a.csv"**  in "read mode" 
   (Download the file from your omnivox account )

    We need to read all the data from the file (file is in csv format and contains a header)

    . We need to verify that the phone column is in the following format :

        ddd-ddd-dddd (where d is a digit) (i.e. 514-321-0909)

    . We need to verify that the email column is in the following format :
        
        xxxxxxx@xxxxxxx.xxx (where x are alphanumeric character)
    
      Use REGEX (Regular expressions to check the records (hint : after reading a row,   you can convert the list record into a string (using str) and then apply the regex to the resulting string. See your class notes for an example. (This is a reasonable step to complete this exercise, although it may not be suitable for production code). 
		
    . All records found with a phone that does not follow the given format should be written to a file : bad_phone_recs.csv

    . All records found with an email that does not follow the given format should be written to a file : bad_email_recs.csv

    . All good records should be written to a file : good_recs.csv
   	


```python
import os, re, csv

# Define general REGEX expressions for bot phone and email
simple_phone_re = '\d{3}-\d{3}-\d{4}'
simple_email_re = '\w+@\w+[.]\w+'

# Open the ca_sample_data_a.csv file in your specific directory
# You can use a try ... exception to catch error when opening the file (optional for this exercise)
try:
    #os.chdir(r'c:\users') #Optional
    filehandle = open('ca_sample_data_a.csv', 'r') 

except:
    print("Error opening dataset ... check the file directory and file name!")
else:
    
    # Create empty lists to hold good and bad records
    bad_phone_recs = []
    good_recs = []
    bad_email_recs = []
    
    # Create csv reader to start reading information
    file_reader = csv.reader(filehandle)

    file_header = next(file_reader, None) #filter header line and store it in a variable for later use
    for row in file_reader:
                   
        # Search for phone and emails
            
        phone_to_verify = row[7]
        email_to_verify = row[8]
                
        #------------------------------------------------------
        match_phone = re.search(simple_phone_re, phone_to_verify) #notice we make the row into a string
        match_email = re.search(simple_email_re, email_to_verify)
            
        # Decide which list to populate
        if match_phone and match_email:
            good_recs.append(row)
            
        else:
            if not(match_phone):
                bad_phone_recs.append(row)
            if not(match_email):
                bad_email_recs.append(row)
    
    #close original input file
    filehandle.close()


# Now open the files that will contain the results
# Write csv files containing good records, bad phone records and bad email records

try:
    #os.chdir(r'c:\users\') #optional
    
    filehandle_good = open('good_recs.csv', 'w', newline='')
    filehandle_phone = open('bad_phone_recs.csv', 'w', newline='') 
    filehandle_email = open('bad_email_recs.csv', 'w', newline='')

except:
    print("Error opening datasets for writing bad records ...!")
else:
     
    # define csv writers for all 3 files
    file_writer_phone = csv.writer(filehandle_phone)
    file_writer_email = csv.writer(filehandle_email)
    file_writer_good = csv.writer(filehandle_good)
    
    # Write the header and the data contents for all 3 files

    file_writer_good.writerow(file_header)
    file_writer_good.writerows(good_recs)

    file_writer_phone.writerow(file_header)
    file_writer_phone.writerows(bad_phone_recs)
    
    file_writer_email.writerow(file_header)
    file_writer_email.writerows(bad_email_recs) 
    
    # Close files
   
    filehandle_good.close()
    filehandle_phone.close()
    filehandle_email.close()

    
               

```
